## Adding Rotten Tomatoes  Movie Info to SQL Data

```python
rt_movie_info = pd.read_csv("../ZippedData/rt.movie_info.tsv.gz",sep='\t', on_bad_lines='skip')
rt_movie_info.head()
```

```python
rt_movie_info.to_sql('rt_movie_info',conn, if_exists='replace')
```

```python
#Checking to make sure table was added
pd.read_sql("""
SELECT name
FROM sqlite_master
WHERE type = "table"

""", conn)
```

```python
# Connect to your SQLite database

try:
    # Create a cursor object
    cursor = conn.cursor()
    
    # Drop the table
    cursor.execute("DROP TABLE IF EXISTS 'rt.movie_info';")  # Use quotes for the table name with a dot
    print("Table 'rt.movie_info' dropped successfully.")
    
    # Commit the changes
    conn.commit()

except sqlite3.Error as e:
    print(f"An error occurred: {e}")

finally:
    # Close the connection
    conn.close()
```
